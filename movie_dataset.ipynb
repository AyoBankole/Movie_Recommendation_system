```python
# ----------mearge each datasets to improve data readability
movie_data= pd.merge(credits, movies, on= 'title', how= 'left')
movie_data.head()
```

```python
movie_data.columns
```

```python
# -------------remove unwanted columns
movie_data.drop(['movie_id', 'crew'], axis= 1, inplace= True)
movie_data.head()
```

```python
# ---------------divide dataset into numerical and cartegorical columns
num = movie_data.select_dtypes(include= 'number')
cat = movie_data.select_dtypes(exclude= 'number')

display(num.head(3))
display(cat.head(3))
```

```python
# -----------------clean the dataset 
for col in movie_data.columns:
    if movie_data[col].isnull().sum()/len(movie_data) * 100 < 30:
        if col in num.columns:
            movie_data[col].fillna(movie_data[col].median(), inplace= True)
        else:
            movie_data[col].fillna(movie_data[col].mode()[0], inplace= True)
    else:
        movie_data.drop(col, axis= 1, inplace= True)

movie_data.isnull().sum()
```

```python
# --------------further drop more columns to streamline dataset to fit output choice
movie_data.drop(['original_language'], axis= 1, inplace= True)
movie_data.columns
```

<hr>It's challenging to process the vocabularies in a handful of our articles. As a result, we will choose and repopulate the top 10 pertinent data entries from the designated column. We will not make any changes to the current dataframe going forward; instead, a new dataframe will be created.

```python
# -----------copy the dataset into a new variable so as to retain the details in the dataset
processed_data = movie_data.copy()
```

```python
# -----------------import the necessary libraries
from sklearn.metrics.pairwise import linear_kernel
from sklearn.feature_extraction.text import TfidfVectorizer

# -----------drop duplicates
processed_data.drop_duplicates(inplace= True)
processed_data.isnull().sum()
```

```python
# --------------drop the "id" column
processed_data.drop('id', axis= 1, inplace= True)
```

```python
# -----------------------import the ast library to extract words literally from the column 
import ast
ast.literal_eval(processed_data["spoken_languages"][0])
```

```python
list(set([i["name"] for i in ast.literal_eval(processed_data["spoken_languages"][0])][:10]))
```

```python
# --------------creating a function that will implement the codes above in order to effect the extraction
def extractor(row):
    row_list = ast.literal_eval(row)
    first_10_unique_names = set([i["name"] for i in row_list][:10])
    first_10 = ", ".join(f"{name.replace(' ', '')}" for name in first_10_unique_names)
    return first_10

# Apply the function to the column containing lists of dictionaries
processed_data['cast'] = processed_data['cast'].apply(extractor)
processed_data['genres'] = processed_data['genres'].apply(extractor)
processed_data['keywords'] = processed_data['keywords'].apply(extractor)
processed_data['spoken_languages'] = processed_data['spoken_languages'].apply(extractor)
processed_data['production_companies'] = processed_data['production_companies'].apply(extractor)
processed_data['production_countries'] = processed_data['production_countries'].apply(extractor)
```

```python
processed_data.head()
```

```python
# -----------------extract the years from the release_dat and rename the columns
processed_data['release_date'] = processed_data['release_date'].str.split('-').str[0]
processed_data = processed_data.rename(columns={'release_date': 'year'})
processed_data.head(3)
```

<b><li>>Visualization of the distribution of movies across different decades using a count plot. and showing the top 10 years with the highest production.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a count plot displaying the year with the highest production
year_release = processed_data['year']  # Assuming 'year' is a column in your DataFrame

# Calculate the count of movies released each year
count_year = year_release.value_counts().sort_index()

# Filter to include only the 20 most recent years
count_year = count_year[-20:]

# Identify the year with the highest production
year_with_highest_production = count_year.idxmax()

# Create a count plot
plt.figure(figsize=(12, 6))
sns.countplot(x='year', data=processed_data, order=count_year.reset_index()['year'].tolist(), color='skyblue')

# Set labels and title
plt.xlabel('Year')
plt.ylabel('Number of Movies Released')
plt.title('Distribution of Movie Releases by Year')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Highlight the year with the highest production
plt.axvline(x=year_with_highest_production, color='red', linestyle='--', linewidth=2, label=f'Highest Production Year: {year_with_highest_production}')

plt.legend()

plt.tight_layout()
plt.show()
```

```python
# --------------using WordCloud to further explore the dataset
# ------------prepare data for visualization
genres=[] #-----------create a list where all genres will be stored
for genre in processed_data.genres:
    
    x=genre.split('|') #--------------split genres by "|"
    for i in x:
         if i not in genres:
            genres.append(str(i))
genres=str(genres)    
titles=[] #------------a list where all movie titles will be stored
for title in processed_data.title:
    titles.append(title[0:-7])
titles=str(titles)
```
